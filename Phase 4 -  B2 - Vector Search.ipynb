```python
%sql
CREATE OR REPLACE TABLE northstar.gold.reviews_for_search AS
SELECT review_id, review_comment_message, sentiment, topic
FROM northstar.gold.review_insights
WHERE review_comment_message IS NOT NULL
LIMIT 2000;                                   -- small = fast sync

ALTER TABLE northstar.gold.reviews_for_search
SET TBLPROPERTIES (delta.enableChangeDataFeed = true);
```

```python
from databricks.sdk import WorkspaceClient

w = WorkspaceClient()
results = w.vector_search_indexes.query_index(
    index_name="northstar.gold.reviews_index",
    columns=["review_id", "review_comment_message", "sentiment", "topic"],
    query_text="damaged or broken products",                      # ← English query!
    num_results=5)

for row in results.result.data_array:
    print(row)
```
