```python
%sql
SELECT
  review_id,
  review_comment_message,
  ai_analyze_sentiment(review_comment_message) AS sentiment
FROM northstar.silver.reviews
WHERE review_comment_message IS NOT NULL
LIMIT 5;
```

```python
%sql
SELECT
  review_id,
  review_comment_message,
  ai_analyze_sentiment(review_comment_message) AS sentiment,
  ai_classify(review_comment_message,
              ARRAY('delivery', 'product quality', 'price', 'other')) AS topic,
  ai_summarize(review_comment_message, 15) AS short_summary
FROM northstar.silver.reviews
WHERE review_comment_message IS NOT NULL
LIMIT 5;
```

```python
%sql
SELECT COUNT(*) AS reviews_with_text
FROM northstar.silver.reviews
WHERE review_comment_message IS NOT NULL;
```

```python
%sql
CREATE OR REPLACE TABLE northstar.gold.review_insights AS
SELECT
  review_id,
  order_id,
  review_score,
  review_comment_message,
  ai_analyze_sentiment(review_comment_message) AS sentiment,
  ai_classify(review_comment_message,
              ARRAY('delivery', 'product quality', 'price', 'other')) AS topic,
  ai_summarize(review_comment_message, 15) AS short_summary
FROM northstar.silver.reviews
WHERE review_comment_message IS NOT NULL;   -- no LIMIT = all reviews with text
```

```python
%sql
SELECT COUNT(*) AS total_insights FROM northstar.gold.review_insights;
```

```python
%sql
SELECT topic, sentiment, COUNT(*) AS n
FROM northstar.gold.review_insights
GROUP BY topic, sentiment
ORDER BY n DESC;
```
